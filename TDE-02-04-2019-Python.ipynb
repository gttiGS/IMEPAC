```python
dataset['size'] = (dataset['Cnes']*20)
```

```python
dataset['size']
```

```python
def categoriza(s):
    if s >= 80:
        return 'Big'
    elif s >= 60:
        return 'Medium'
    elif s >= 40:
        return 'Small'
```

```python
dataset['cat_size'] = dataset['size'].apply(categoriza)
```

```python
dataset['cat_size']
```

```python
pd.value_counts(dataset['cat_size'])
```

```python
dataset.drop(['cat_size'], axis=1, inplace=True)
```

```python
dataset.drop(['size'], axis=1, inplace=True)
```

```python
dataset.drop(dataset[dataset.Cnes==0].index ,inplace=True)
```

```python
dataset.drop(dataset[dataset.Cnes>0].index ,inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.dropna(how='all', inplace=True)
```

```python
dataset['Cnes'].fillna(dataset['Cnes'].mean(), inplace=True)
```

```python
dataset['Cnpj'].fillna(1, inplace=True)
```
